```python
# ğŸ”¹ 1. Cargar los dos CSVs de posiciones en espaÃ±ol
df_original_sp = pd.read_csv('posiciones_ciudades_originales.csv')
df_corregidas_sp = pd.read_csv('posiciones_ciudades_corregidas.csv')

# ğŸ”¹ 2. FunciÃ³n para limpiar las ciudades
def limpiar_palabra(ciudad):
    return ciudad.rstrip(',').strip()

# ğŸ”¹ 3. Limpiar ambas tablas
df_original_sp['palabra'] = df_original_sp['palabra'].apply(limpiar_palabra)
df_corregidas_sp['palabra'] = df_corregidas_sp['palabra'].apply(limpiar_palabra)

# ğŸ”¹ 4. Convertir 'posiciÃ³n' a numÃ©rico
df_original_sp['posiciÃ³n'] = pd.to_numeric(df_original_sp['posiciÃ³n'], errors='coerce')
df_corregidas_sp['posiciÃ³n'] = pd.to_numeric(df_corregidas_sp['posiciÃ³n'], errors='coerce')

# ğŸ”¹ 5. Concatenar y ordenar
df_final_sp = pd.concat([df_original_sp, df_corregidas_sp]).sort_values(by=['palabra', 'posiciÃ³n'])
df_final_sp = df_final_sp.drop_duplicates(subset=['palabra'], keep='first')

# ğŸ”¹ 6. Orden final
df_final_sp = df_final_sp.sort_values(by='posiciÃ³n', na_position='last')
df_final_sp['posiciÃ³n'] = df_final_sp['posiciÃ³n'].apply(lambda x: 'none' if pd.isna(x) else int(x))

# ğŸ”¹ 7. Merge limpio (evitamos duplicados)
if 'sp_posiciones' in df_cities.columns:
    df_cities = df_cities.drop(columns=['sp_posiciones'])

df_cities = df_cities.merge(
    df_final_sp.rename(columns={'palabra': 'city_in_Spanish', 'posiciÃ³n': 'sp_posiciones'}), 
    on='city_in_Spanish', 
    how='left'
)

# ğŸ”¹ 8. Verificar faltantes
missing_sp = df_cities[df_cities['sp_posiciones'].isin(['none', 0]) | df_cities['sp_posiciones'].isna()]

# ğŸ”¹ 9. Imprimir TODO el DataFrame para ver cÃ³mo quedÃ³ la tabla completa
print("\nâœ… DataFrame completo con 'sp_posiciones' agregado:\n")
print(df_cities.head(20))  # si quieres ver mÃ¡s filas cambia el 20 por un nÃºmero mayor

print(f"\nâš ï¸ Ciudades en espaÃ±ol sin posiciÃ³n asignada (total: {len(missing_sp)}):\n")
print(missing_sp[['city_in_Spanish', 'sp_posiciones']])
```

```python
# ğŸ”¹ 1. Cargar los dos CSVs de posiciones en indonesio
df_original_id = pd.read_csv('posiciones_ciudades_ID.csv')
df_corregidas_id = pd.read_csv('posiciones_ciudades_corregidas_ID.csv')

# ğŸ”¹ 2. Convertir la columna 'posiciÃ³n' a numÃ©rico
df_original_id['posiciÃ³n'] = pd.to_numeric(df_original_id['posiciÃ³n'], errors='coerce')
df_corregidas_id['posiciÃ³n'] = pd.to_numeric(df_corregidas_id['posiciÃ³n'], errors='coerce')

# ğŸ”¹ 3. Concatenar y ordenar
df_final_id = pd.concat([df_original_id, df_corregidas_id]).sort_values(by=['palabra', 'posiciÃ³n'])
df_final_id = df_final_id.drop_duplicates(subset=['palabra'], keep='first')

# ğŸ”¹ 4. Orden final
df_final_id = df_final_id.sort_values(by='posiciÃ³n', na_position='last')
df_final_id['posiciÃ³n'] = df_final_id['posiciÃ³n'].apply(lambda x: 'none' if pd.isna(x) else int(x))

# ğŸ”¹ 5. Definir la funciÃ³n de limpieza para las ciudades en indonesio
def remover_kota_y_bahasa(nombre_ciudad):
    if isinstance(nombre_ciudad, str):
        if nombre_ciudad.lower().startswith("kota "):
            nombre_ciudad = nombre_ciudad[5:]
        palabras_a_eliminar = ["Bahasa Indonesia:", "Bahasa", "Bahasa Indonesia", "Provinsi"]
        for palabra in palabras_a_eliminar:
            if nombre_ciudad.startswith(palabra):
                nombre_ciudad = nombre_ciudad[len(palabra):].strip()
    return nombre_ciudad.strip()

# ğŸ”¹ 6. Crear la columna auxiliar en df_cities
df_cities['city_in_Indonesian_clean'] = df_cities['city_in_Indonesian'].apply(remover_kota_y_bahasa)

# ğŸ”¹ 7. Merge usando la columna "clean"
if 'id_posiciones' in df_cities.columns:
    df_cities = df_cities.drop(columns=['id_posiciones'])

df_cities = df_cities.merge(
    df_final_id.rename(columns={'palabra': 'city_in_Indonesian_clean', 'posiciÃ³n': 'id_posiciones'}),
    on='city_in_Indonesian_clean', 
    how='left'
)

# ğŸ”¹ 8. Verificar faltantes
missing_id = df_cities[df_cities['id_posiciones'].isin(['none', 0]) | df_cities['id_posiciones'].isna()]

# ğŸ”¹ 9. Imprimir TODO el DataFrame completo para validar
print("\nâœ… DataFrame completo con 'id_posiciones' agregado:\n")
print(df_cities.head(20))

print(f"\nâš ï¸ Ciudades en indonesio sin posiciÃ³n asignada (total: {len(missing_id)}):\n")
print(missing_id[['city_in_Indonesian', 'city_in_Indonesian_clean', 'id_posiciones']])
```
