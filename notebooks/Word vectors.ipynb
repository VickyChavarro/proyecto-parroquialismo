```python
import pandas as pd

# Cargar el archivo CSV con las traducciones a portugu√©s
df = pd.read_csv(r'C:\Users\victo\Downloads\enriched-world-city-listing-rus-por-fr.csv')

# Extraer las ciudades en portugu√©s, eliminando duplicados y NaN
ciudades_portugues = df['city_in_Portuguese'].dropna().unique()

# Obtener el vocabulario del modelo en portugu√©s
vocabulario_pt = modelo_fasttext_pt.get_words()

# Crear un diccionario de palabra ‚Üí posici√≥n en el vocabulario
vocabulario_con_posiciones_pt = {palabra: idx for idx, palabra in enumerate(vocabulario_pt)}

# Buscar cada ciudad en el vocabulario portugu√©s y guardar la posici√≥n
resultados_pt = {}
contador_none_pt = 0

for ciudad in ciudades_portugues:
    posicion = vocabulario_con_posiciones_pt.get(ciudad, None)
    resultados_pt[ciudad] = posicion
    if posicion is None:
        contador_none_pt += 1
    print(f"La posici√≥n de la palabra '{ciudad}' es {posicion}")

# Mostrar resumen
print(f"\nTotal de ciudades en portugu√©s: {len(ciudades_portugues)}")
print(f"Ciudades no encontradas en el vocabulario: {contador_none_pt}")
```

```python
import pandas as pd

# üîπ 1. Convertir el diccionario de resultados en una lista limpia
datos_pt = []
for palabra, posicion in resultados_pt.items():
    if isinstance(posicion, (int, float)):
        datos_pt.append([palabra.strip(), int(posicion)])
    else:
        datos_pt.append([palabra.strip(), 'none'])

# üîπ 2. Crear el DataFrame
df_resultados_pt = pd.DataFrame(datos_pt, columns=['palabra', 'posici√≥n'])

# üîπ 3. Guardar en CSV con codificaci√≥n UTF-8 (para acentos en portugu√©s)
df_resultados_pt.to_csv('posiciones_ciudades_PT.csv', index=False, encoding='utf-8-sig')

print("üìÇ Archivo 'posiciones_ciudades_PT.csv' guardado correctamente.")
```

```python
# Ordenar los resultados por posici√≥n, poniendo las ciudades no encontradas al final
resultados_ordenados_pt = sorted(resultados_pt.items(), key=lambda x: (x[1] is None, x[1]))

# Imprimir las ciudades ordenadas desde la m√°s frecuente a la menos frecuente
print("Ciudades en portugu√©s ordenadas desde la m√°s frecuente a la menos frecuente:")
for ciudad, posicion in resultados_ordenados_pt:
    if posicion is None:
        print(f"La ciudad '{ciudad}' no fue encontrada en el vocabulario.")
    else:
        print(f"La ciudad '{ciudad}' est√° en la posici√≥n {posicion}.")
```

```python
import pandas as pd

# Convertir el diccionario a un DataFrame
df_resultados_en = pd.DataFrame(resultados_en.items(), columns=['palabra', 'posici√≥n'])

# üîπ 1. Reemplazar valores None por 'none' para mantener consistencia
df_resultados_en['posici√≥n'] = df_resultados_en['posici√≥n'].apply(lambda x: 'none' if x is None else x)

# üîπ 2. Guardar en CSV sin √≠ndices
df_resultados_en.to_csv('posiciones_ciudades_EN.csv', index=False)

print("üìÇ Archivo 'posiciones_ciudades_EN.csv' guardado correctamente.")
print(df_resultados_en)
```
